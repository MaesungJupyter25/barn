```python
# 이건 해당 코랩 파일의 기초적인 종속성을 불러오는 코드입니다.
# 그냥 실행 한번 시켜주세요.
# 다른 코랩 파일에도 여럿 볼 수 있을겁니다. 적당히 눈치채고 실행시키세요.

import numpy as np
import matplotlib.pyplot as plt

print("종속성 불러오기 완료")
```

```python
# 이것또한 마찬가지
def draw_func(func):
  x = np.arange(1, 100, 0.05)
  y = [func(x) for x in x]
  plt.plot(x, y)

print("기능 불러오기 완료")
```
